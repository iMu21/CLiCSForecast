```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

file_path = "../data_analysis_reports/Sample_Data.xlsx"
data = pd.read_excel(file_path)

# Splitting the data into features (X) and target (y)
X = data.drop(columns=['Month', 'Year', 'Payment Amount'])  # Adjust the column names if necessary
y = data['Payment Amount']

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fitting a linear regression model
model = LinearRegression()
model.fit(X_scaled, y)

# Getting the coefficients of the model
coefficients = model.coef_
feature_names = X.columns

# Creating a DataFrame to display the relation with the payment amount
relation_df = pd.DataFrame({
    'Field Name': feature_names,
    'Relation with Payment (%)': np.abs(coefficients) / np.sum(np.abs(coefficients)) * 100
})

# Sorting the DataFrame by the relation with payment
relation_df_sorted = relation_df.sort_values(by='Relation with Payment (%)', ascending=False)

# Display the sorted DataFrame
print(relation_df_sorted)
```

```python
import pandas as pd

# Drop the 'Month' and 'Year' columns as they are not needed for correlation calculation
data_numeric = data.drop(columns=['Month', 'Year'])

# Calculating the correlation matrix
correlation_matrix = data_numeric.corr()

# Extracting the correlation of each field with the 'Payment Amount'
payment_correlation = correlation_matrix['Payment Amount'].drop('Payment Amount')

# Creating a DataFrame to display the correlation with the payment amount
correlation_df = pd.DataFrame({
    'Field Name': payment_correlation.index,
    'Correlation with Payment Amount': payment_correlation.values
})

# Sorting the DataFrame by the correlation with payment amount
correlation_df_sorted = correlation_df.sort_values(by='Correlation with Payment Amount', ascending=False)

# Display the sorted DataFrame
print(correlation_df_sorted)
```

```python
import pandas as pd

data_numeric = data.drop(columns=['Month', 'Year'])

# Calculating the correlation matrix
correlation_matrix = data_numeric.corr()

# Extracting the correlation of each field with the 'Payment Amount'
payment_correlation = correlation_matrix['Payment Amount'].drop('Payment Amount')

# Creating a dictionary to store the correlations
correlation_dict = payment_correlation.to_dict()

# Defining the object with each field as a property
class CorrelationWithPayment:
    def __init__(self, correlations):
        for field, value in correlations.items():
            setattr(self, field, value)

# Creating an instance of the object
correlation_object = CorrelationWithPayment(correlation_dict)

# Printing the properties of the object
for field in correlation_dict.keys():
    print(f'{field}: {getattr(correlation_object, field)}')
```
