```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data_analysis_reports/DataAnalaysisReportClaimForecast.csv")
```

```python
data
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

data = data.drop(['year'],axis=1)
data = data.drop(['month'],axis=1)
data = data.drop(['enroll'],axis=1)
data = data.drop(['enroll_product'],axis=1)
data = data.drop(['dependent'],axis=1)

x = data.drop(['claim_amount'],axis=1)
y = data['claim_amount']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.corr()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train, y_train = train_data.drop(['claim_amount'], axis = 1),train_data['claim_amount']
x_train_s = scaler.fit_transform(x_train)

reg = LinearRegression()
reg.fit(x_train_s,y_train)
```

```python
test_data = x_test.join(y_test)
```

```python
x_test, y_test = test_data.drop(['claim_amount'],axis=1),test_data['claim_amount']
```

```python
x_test_s = scaler.transform(x_test)
```

```python
reg.score(x_test_s,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train_s,y_train)
```

```python
forest.score(x_test_s,y_test)
```
